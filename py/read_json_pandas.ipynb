```python
import json
from pandas.io.json import json_normalize
```

```python
def load_json_file(filename):
    with open(filename) as json_data:
        list_data = json.load(json_data)
    return list_data
```

```python
d1 = load_json_file('materials1.json')
d2 = load_json_file('materials2.json')
d3 = load_json_file('materials3.json')
d4 = load_json_file('materials4.json')
d5 = load_json_file('materials5.json')
```

```python
d_comb = d1 + d2 + d3 + d4 + d5
```

```python
df = json_normalize(d_comb)
```

```python
for index, row in df.iterrows():
    
    chemicalFormula = row['sample.material.chemicalFormula']
    mpID = row['sample.material.condition'][0]['scalar']
    volume = row['sample.material.condition'][1]['scalar']
    volumeUnits = row['sample.material.condition'][1]['units']
    formationEnergy = row['sample.measurement'][0]['property']['scalar']
    formationEnergyUnits = row['sample.measurement'][0]['property']['units']
    bandGap = row['sample.measurement'][1]['property']['scalar']
    bandGapUnits = row['sample.measurement'][1]['property']['units']
    density = row['sample.measurement'][2]['property']['scalar']
    densityUnits = row['sample.measurement'][2]['property']['units']
    
    df.loc[index,'chemicalFormula'] = chemicalFormula
    df.loc[index,'mpID'] = mpID
    df.loc[index,'volume'] = volume
    df.loc[index,'volumeUnits'] = volumeUnits
    df.loc[index,'spaceGroup'] = spaceGroup
    df.loc[index,'formationEnergy'] = formationEnergy
    df.loc[index,'formationEnergyUnits'] = formationEnergyUnits
    df.loc[index,'bandGap'] = bandGap
    df.loc[index,'bandGapUnits'] = bandGapUnits
    df.loc[index,'density'] = density
    df.loc[index,'densityUnits'] = densityUnits
```

```python
# Add Crystal system
for index, row in df[df['bandGap'].between(0.5, 2.5)].iterrows():
    if len(row['sample.material.condition']) > 3:
        if row['sample.material.condition'][3]['name'] == 'Crystal system':
            crystalSystem = row['sample.material.condition'][3]['scalar']
            df.loc[index,'crystalSystem'] = crystalSystem
    if len(row['sample.material.condition']) > 4:
        if row['sample.material.condition'][4]['name'] == 'Crystal system':
            crystalSystem = row['sample.material.condition'][4]['scalar']
            df.loc[index,'crystalSystem'] = crystalSystem
```

```python
df[df['crystalSystem'].notnull()]
```

```python
df_export = df[df['crystalSystem'].notnull()]
```

```python
df_export = df_export.drop(['sample.material.chemicalFormula','sample.material.condition','sample.measurement','sample.reference'], 1)
```

```python
df_export = df_export.reset_index(drop = True)
```

```python
# Check if all units are the same
# If no output, then all units are the same
for index, row in df_export.iterrows():
    if (row['volumeUnits'] != '$\AA^3$'):
        print('Index {0}: volumeUnits = {1}'.format(index, row['volumeUnits']))
    if (row['formationEnergyUnits'] != 'eV/atom'):
        print('Index {0}: formationEnergyUnits = {1}'.format(index, row['formationEnergyUnits']))
    if (row['bandGapUnits'] != 'eV'):
        print('Index {0}: bandGapUnits = {1}'.format(index, row['bandGapUnits']))
    if (row['densityUnits'] != 'g/cm$^3$'):
        print('Index {0}: densityUnits = {1}'.format(index, row['densityUnits']))
```
